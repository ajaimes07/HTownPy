```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
school_data = "./school_data.csv"
school_df = pd.read_csv(school_data)
school_df.head()
```

```python
#transform school dataframe
school_df.drop(['OBJECTID_12', 'OBJECTID_1','OBJECTID', 'DISTRICT','CAMPUS',
                'CITY_MAIL','ZIP_MAIL','STREET_MAI','COUNTY',
                'REGION','GRADERANGE','GRADEGRP', 'VERIFY','INSTR_TYPE',
               'STREET_M_1', 'ADJUST', 'Shape'], axis=1, inplace=True)
school_df.head()
```

```python
school_cols = ["DISTNAME","CAMPNAME","STREET","CITY","ZIP","CNTYNAME"]
school_transformed = school_df[school_cols].copy()

# Rename the column headers
school_transformed = school_transformed.rename(columns={"DISTNAME": "DISTRICT_NAME",
                                                         "CAMPNAME": "CAMPUS_NAME",
                                                         "CNTYNAME": "COUNTY_NAME",
                                                         })

# Set index
school_transformed.set_index("DISTRICT_NAME", inplace=True)

school_transformed.head()

#school_df.to_csv (r'C:\Desktop\export_dataframe.csv', index = None, header=True)
```

```python
connection_string = "koyatomi:Tomisin30@localhost:5432/school_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
school_transformed.to_sql(name='school', con=engine, if_exists='append', index=True)
```
