```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import string as string
csv_path = "Desktop/DC_Registration_All.csv"
dryclean_df = pd.read_csv(csv_path)
dryclean_df.head()

```

```python
#dropping columns that wont be used to tell us anything
dryclean_df = dryclean_df[["REG_ENT_NAME", "CITY_NAME", "ZIP_CD", "COUNTY", "PHYS_NAME"]]
dryclean_df.head()
```

```python
#Output cleaned data to csv
#housepy_df.to_csv("housepy_cleaned.csv", index=False, header=True)
```

```python
#Clean data for no value cells

dryclean_df.dtypes

dryclean_df.dropna(inplace=True)
```

```python
#df.a = df.a.astype(float)

dryclean_df['ZIP_CD'] = dryclean_df['ZIP_CD'].astype(int)
dryclean_df.head()
```

```python
dryclean_df['COUNTY'] = dryclean_df['COUNTY'].astype(str)
```

```python

dryclean_df.dtypes
```

```python
DRYCLEAN1 = dryclean_df[dryclean_df["COUNTY"]=="BRAZORIA"]

DRYCLEAN1.reset_index(inplace=True)

DRYCLEAN1.head()
```

```python
DRYCLEAN2 = dryclean_df[dryclean_df["COUNTY"]=="BRAZOS"]

DRYCLEAN2.reset_index(inplace=True)

DRYCLEAN2.head()
```

```python
DRYCLEAN3 = dryclean_df[dryclean_df["COUNTY"]=="FORT BEND"]

DRYCLEAN3.reset_index(inplace=True)

DRYCLEAN3.head()
```

```python
DRYCLEAN4 = dryclean_df[dryclean_df["COUNTY"]=="GALVESTON"]

DRYCLEAN4.reset_index(inplace=True)

DRYCLEAN4.head()
```

```python
DRYCLEAN5 = dryclean_df[dryclean_df["COUNTY"]=="HARRIS"]

DRYCLEAN5.reset_index(inplace=True)

DRYCLEAN5.head()
```

```python
DRYCLEAN6 = dryclean_df[dryclean_df["COUNTY"]=="MONTGOMERY"]

DRYCLEAN6.reset_index(inplace=True)

DRYCLEAN6.head()
```

```python
#Rmerge_math_score = pd.merge(Rninth_grade_df, Rtenth_grade_df, on="school_name")
#pd.merge(df1, df2, on = ['Name', 'Parent', 'Parent_Addr'], how = 'outer')

merge1 = pd.merge(DRYCLEAN1, DRYCLEAN2, on = ['REG_ENT_NAME', 'CITY_NAME', 'ZIP_CD', 'COUNTY', 'PHYS_NAME'], how= 'outer')
merge1.head()
```

```python
merge2 = pd.merge(merge1, DRYCLEAN3, on = ['REG_ENT_NAME', 'CITY_NAME', 'ZIP_CD', 'COUNTY', 'PHYS_NAME'], how= 'outer')
merge2.head()
```

```python
merge3 = pd.merge(merge2, DRYCLEAN4, on = ['REG_ENT_NAME', 'CITY_NAME', 'ZIP_CD', 'COUNTY', 'PHYS_NAME'], how= 'outer')
merge3.head()
```

```python
merge4 = pd.merge(merge3, DRYCLEAN5, on = ['REG_ENT_NAME', 'CITY_NAME', 'ZIP_CD', 'COUNTY', 'PHYS_NAME'], how= 'outer')
merge4.head()
```

```python
merge5 = pd.merge(merge4, DRYCLEAN6, on = ['REG_ENT_NAME', 'CITY_NAME', 'ZIP_CD', 'COUNTY', 'PHYS_NAME'], how= 'outer')
merge5.head()
```

```python
merge5 = merge5[["REG_ENT_NAME", "CITY_NAME", "ZIP_CD", "COUNTY", "PHYS_NAME"]]
merge5.head()
```

```python
merge5.to_csv('Dryclean_project.csv', index=False, header=True)
```
